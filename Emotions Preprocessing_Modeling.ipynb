```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize a CountVectorizer object
count_vectorizer = CountVectorizer(stop_words = 'english',ngram_range = (1,2))

# Fit and transform the training data  
count_train = count_vectorizer.fit_transform(X_train.values.astype('U'))

# Transform the test data 
count_test = count_vectorizer.transform(X_test.values.astype('U'))

# Print the first 10 features of the count_vectorizer
print(count_vectorizer.get_feature_names_out()[:-10])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize a TfidfVectorizer object
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the training data using tfidf
tfidf_train = tfidf_vectorizer.fit_transform(X_train.values.astype('U'))

# Transform the test data 
tfidf_test = tfidf_vectorizer.transform(X_test.values.astype('U'))

# Print the first 10 features of the tfidf_vectorizer
print(tfidf_vectorizer.get_feature_names_out()[:-10])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import classification_report



def create_pipeline(classifier):
    #Helper function to create pipeline for each model
    
    pipeline = Pipeline([('classifier', OneVsRestClassifier(classifier))],verbose = True)
    return pipeline
```

```python
# Create function to standardize process of creating instance of model, fitting and predicting and creating
# classification report

def train_model(pipeline,X_train = count_train,y_train = y_train, X_test = count_test,y_test = y_test):

    print("Training the model...")
    with tqdm(total=1, desc="Training", leave=False, bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt} {elapsed}') as pbar:
    
        pipeline.fit(X_train, y_train)
        pbar.update(1)
        
    print("Evaluating the model...")
    y_pred = pipeline.predict(X_test)
    
    print(f"This is the classification report for {pipeline.named_steps}")
    print(classification_report(y_test, y_pred, target_names=emotions,zero_division = np.nan))
```

In this project, the goal of the emotion identification is to be used for a chatbot for the purpose of crisis text/therapy etc. For this purpose, it is more important that the model is correctly identifying an emotion when it labels one rather than catching every instance of every emotion. For example, it is ok if the chatbot does not recognize every instance of someone saying something that indicates sadness. However, it is important that the chatbot does not mistake sadness for happiness. So, precision is the primary metric to be maximized in this model while still having decent recall and ability to identify different emotions. 

Looking at the different models here, Random Forest provides the best precision while still identifying most of the emotions. While Naive Bayes has a higher precision in theory, it does so by simply ignoring most of the emotions (shown by the very low recall number). The advantage of random forest is that it can also continue to be adapted for more emotions with appropriately scaling complexity.
